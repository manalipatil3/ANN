```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
```

```python
(X_train,y_train),(X_test,y_test)=mnist.load_data()
```

```python
X_train.shape
```

```python
X_train[0]
```

```python
y_train[0]
```

```python
X_train=X_train/255.0
X_test=X_test/255.0
```

```python
X_train.shape
```

```python
model_no_hidden=Sequential([
    Flatten(input_shape=(28,28)),
    Dense(10,activation='softmax')
])
```

```python
model_no_hidden.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model_no_hidden.fit(X_train,y_train,epochs=5,batch_size=32,validation_split=0.2,verbose=1)
```

```python

test_loss_no_hidden, test_acc_no_hidden=model_no_hidden.evaluate(X_test,y_test)
```

```python
y_pred=model_no_hidden.predict(X_test)

```

```python
print(y_pred)
```

```python
y_pred_labels=y_pred.argmax(axis=1)
```

```python
print(y_pred_labels)
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred_labels)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model_with_hidden=Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128,activation='relu'),
    Dense(10,activation='softmax')
])
```

```python
model_with_hidden.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model_with_hidden.fit(X_train,y_train,epochs=5,batch_size=32,validation_split=0.2,verbose=1)
```

```python
test_loss_hidden, test_acc_hidden=model_with_hidden.evaluate(X_test,y_test)
```

```python
y_pred=model_with_hidden.predict(X_test)
```

```python
y_pred_labels=y_pred.argmax(axis=1)
```

```python
print(y_pred_labels)
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred_labels)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('actual')
```

```python
print("Model Comparison")
print("-------------------------")
print("No Hidden Layer Accuracy:", test_acc_no_hidden)
print("With Hidden Layer Accuracy:", test_acc_hidden)
```

The model with no hidden layer learns only simple linear patterns, so its accuracy is lower.
The model with hidden layer + ReLU learns more complex patterns from digit images, so it performs better on MNIST.
